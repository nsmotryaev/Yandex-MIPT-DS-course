```python
import math
import numpy as np
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
```

```python
%matplotlib inline
from matplotlib import pylab as plt
x = np.arange(0, 16, 0.1)
y = list(map(f,x))
plt.plot(x, y)
plt.show()
```

```python
def f2(x):
    return 3.32512949 - 0.06531159*x - 0.00760104*x**2
```

```python
%matplotlib inline
from matplotlib import pylab as plt
x = np.arange(0, 16, 0.1)
y = list(map(f,x))
y1 = list(map(f1,x))
y2 = list(map(f2,x))
plt.plot(x, y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x1 = 1
x2 = 4
x3 = 10
x4 = 15
an_x1 = f(x1)
an_x2 = f(x2)
an_x3 = f(x3)
an_x4 = f(x4)
a3 = np.array([[x1**0, x1**1, x1**2, x1**3],
              [x2**0, x2**1, x2**2, x2**3],
              [x3**0, x3**1, x3**2, x3**3],
              [x4**0, x4**1, x4**2, x4**3]])
b3 = np.array([an_x1, an_x2, an_x3, an_x4])
k = np.linalg.solve(a3, b3)
def f3(x):
    return k[0] * x**0 + (k[1] * x**1) + (k[2] * x**2) + (k[3] * x**3)
%matplotlib inline
from matplotlib import pylab as plt
x = np.arange(0, 16, 0.1)
y = list(map(f,x))
y1 = list(map(f1,x))
y2 = list(map(f2,x))
y3 = list(map(f3,x))
plt.plot(x, y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```
