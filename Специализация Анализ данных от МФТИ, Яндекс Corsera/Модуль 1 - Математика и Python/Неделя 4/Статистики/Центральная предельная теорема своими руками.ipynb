## 1. Сравниваем гистограмму выборки объемом 1000 экспоненциального случайного распределения с теоретическим графиком плотности распределения.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt   #Импорт необходимых библиотек
import scipy.stats as sts
%matplotlib inline
```

```python
L = 2 #Задаем лямбду - основной параметр экспоненциального распределения
Scale = 1/L # Так как у scipy.stats дерьмовая документация, то может быть неочевидным, 
            #что для задания нужна не Лямбда белого человека, а мат.ожидание 
expon_rv = sts.expon(loc = 0, scale = Scale) #Задаем экспоненциальное распределение с параметром Scale,
                                             #Причем сначала необходимо указать параметр сдвига функции
```

```python
sample = expon_rv.rvs(1000) #создаем выборку объемом 1000
plt.hist(sample, density=True, bins=50, label = 'practical x') #Строим гистограмму
x = np.linspace(0,4,100) # Задаем размеры нашего полотна для графиков
pdf = expon_rv.pdf(x) # Определяем теоретическую эксп. функцию
plt.ylabel('fraction of samples') #Подписываем оси
plt.xlabel('$x$') #Подписываем оси
plt.plot(x, pdf, label='theoretical exp', alpha=0.5 ) #Строим теоретическую плотность распределения
plt.legend() # Для вывода легенды
```

## 2.Сравнение гистограммы распределений выборочных средних с графиком плотности нормального распределения с теми же параметрами.

### а)При объеме выборки 5

```python
n = 5 #Задаем объем выборки
sample_mean_arr = [] #создаем массив для хранения средних значений выборок
for i in range(1000):
    sample = expon_rv.rvs(n) # создание выборки нашего экспоненциального рапределения
    sample_mean_arr.append(sum(sample)/n) # добавляем к массиву среднее значение выборки
plt.hist(sample_mean_arr, density=True, bins=20, label = 'means')
x = np.linspace(0,2,200)
norm = sts.norm(loc = 1/L, scale = np.sqrt((1/L**2)/n))
plt.plot(x, norm.pdf(x), label='theoretical norm' ) #Строим теоретическую плотность распределения
plt.ylabel('fraction of samples') #Подписываем оси
plt.xlabel('$x$') #Подписываем оси
plt.legend() # Для вывода легенды
```
