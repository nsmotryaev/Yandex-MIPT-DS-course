# Центральная предельная теорема своими руками

```python
#Импортируем необходимые библиотеки
import math
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
```

Задаем бета-распределение с параметрами $\alpha=2$ и $\beta=8$.

```python
#Генерация выборки размера 1000 с параметрами alpha и beta
alpha = 2
beta = 8
beta_rv = sts.beta(alpha, beta)
sample = beta_rv.rvs(1000)
```

```python
#Построение гистограммы и теоретической плотности распределения исходного распределения
x = np.linspace(-0.2, 0.8, 100)
pdf = beta_rv.pdf(x)
plt.plot(x, pdf, label='theoretical PDF')

plt.hist(sample, bins = 15, density=True)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

Математическое ожидание и дисперсия бета-распределения с параметрами $\alpha$ и $\beta$ вычисляются по формулам:

$$E(X) = \frac{\alpha}{\alpha+\beta}$$

$$D(X) = \frac{\alpha \beta}{(\alpha+\beta)^2 (\alpha+\beta+1)}$$

```python
#Вычисление теоретических математического ожидания и дисперсии
math_expectation = alpha / (alpha + beta)
var = (alpha * beta) / ((alpha + beta) ** 2 * (alpha + beta + 1))
```

```python
#Генерация 1000 выборок размера n = 5 и вычисление их средних
n = 5
Means = []
for i in range(1000):
    sample_0 = sts.beta(alpha, beta).rvs(n)
    mean = sum(sample_0) / n
    Means.append(mean)
```

Согласно центральной предельной теореме распределение выборочных средних приближается нормальным распределением с параметрами $\mu$ и $\sigma$, где:

$$\mu = E(X) = \frac{\alpha}{\alpha+\beta}$$

$$\sigma^2 = \frac{D(X)}{n} = \frac{\alpha \beta}{n (\alpha+\beta)^2 (\alpha+\beta+1)}$$

```python
#Построение гистограммы распределений выборочных средних и теоретической плотности нормального распределения
x = np.linspace(0, 0.5, 100)
pdf = sts.norm(math_expectation, (var / n) ** 0.5).pdf(x)
plt.plot(x, pdf, label='theoretical PDF')

plt.hist(Means, bins=15, density=True)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

По графику видно, что распределение выборочных средних, действительно, приближается нормальным распределением с параметрами $\mu$ и $\sigma$.

```python
#Генерация 1000 выборок размера n = 10 и вычисление их средних
n = 10
Means = []
for i in range(1000):
    sample_0 = sts.beta(alpha, beta).rvs(n)
    mean = sum(sample_0) / n
    Means.append(mean)
```

```python
#Построение гистограммы распределений выборочных средних и теоретической плотности нормального распределения
x = np.linspace(0, 0.4, 100)
pdf = sts.norm(math_expectation, (var / n) ** 0.5).pdf(x)
plt.plot(x, pdf, label='theoretical PDF')

plt.hist(Means, bins=15, density=True)

#plt.ylabel('$f(x)$')
#plt.xlabel('$x$')
#plt.legend(loc='upper left')
```
