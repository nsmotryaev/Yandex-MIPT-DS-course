```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
expon_rv = sts.expon(0, 1)
sample = expon_rv.rvs(1000)
x = np.linspace(-1,4,100)
pdf = expon_rv.pdf(x)
```

```python
plt.hist(sample, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
```

```python
my_range = []
for k in range(1000):
    my_range.append(expon_rv.rvs(5))
for k in range(1000): 
    my_range[k] = sum(my_range[k])/5
plt.hist(my_range, density=True)
```
