```python
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
```

***Воспользуемся экспоненциальным распределением для того чтобы создать не похожее на нормальное распределение***

```python
expon_distribution = sts.expon()
expon_samples = expon_distribution.rvs(1000) # Выборка из 1000 распределений
```

```python
x = np.linspace(0, 5, 1000)
pdf_expon = expon_distribution.pdf(x) # Расчитываем теор. плотность распределения
```

***Построим гистограмму и наложим теор. плотность распределения***

```python
plt.plot(x, pdf_expon)
plt.hist(expon_samples, density=True)
plt.title('Гистограмма с плотностью распределения')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
def get_mean(n):
    """Получение средних и нормальных распределений на их основе."""
    expon_means = [np.mean(expon_distribution.rvs(n)) for _ in range(1000)]
    # как мы знаем в нормальном распределении
    # mu = EX = lambda - равна матожиданию
    # а значит что sigma = sqrt(lambda / n) - корень (матожидание деленое на кол-во выборок)
    norm_expon = sts.norm(loc=1, scale=np.sqrt(1/n))
    x = np.linspace(0, 3, 1000)
    pdf_expon_norm = norm_expon.pdf(x)
    return x, pdf_expon_norm, expon_means

def draw_graph(x, pdf_norm, expon_means, n):
    """Отприсовка графиков с их теор. плотностью."""
    plt.plot(x, pdf_norm)
    plt.hist(expon_means, density=True)
    plt.title(f'Quantity = {n}')
    plt.xlabel('Mean x')
    plt.ylabel('Mean_x_function')
    plt.show()
```

```python
for n in range(1, 102, 15):
    x, pdf_expon_norm, expon_means = get_mean(n)
    draw_graph(x, pdf_expon_norm, expon_means, n)
```

***Вывод: Чем больше выборки тем больше она похожа на нормальное распределение.  
При n=1 можно заметить что график выглядит также как и при тестируемом запуске.  
При n=16 и n=31 можно заметить что она уже стала похожей на нормальное распределение.  
С каждой итерацией и выбором большего значения выборки, точность аппроксимации распределения выборочных средних нормальным с ростом n увеличивалась***
